## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

ML beats a fixed rule here because the patterns associated with spam are constantly evolving and are often too complex, subtle, and numerous to capture with simple `if-statement` logic. A fixed rule, for instance, might block emails with the word 'prize', but spammers would quickly adapt. Conversely, legitimate emails might also contain 'prize' in specific contexts.

ML models can:
*   **Adapt to new patterns:** Spammers frequently change their tactics. An ML model can be retrained with new data to identify emerging spam characteristics.
*   **Handle high-dimensional data:** Emails have many features (words, sender reputation, links, attachments, formatting, etc.) that interact in complex ways. Fixed rules struggle to combine these effectively.
*   **Learn nuanced relationships:** A specific word might not be spam on its own, but combined with a suspicious sender and multiple links, it becomes highly indicative of spam. ML can learn these complex, non-linear relationships that are impossible to hardcode.
